```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
opposition = "france"
# opposition = "senegal"
#get raw data
passing_network = pd.read_csv(filepath_or_buffer=rf"..\data\england_passing_data_vs_{opposition}.csv")
passing_network
```

```python
#sum up the total pass each person does
passing_network["total_passes"] = passing_network.sum(axis='columns', numeric_only=True)
passing_network
```

```python
#perform arithmetic on the absolute passing statistics and persist the passes
#from and to column

passing_network_as_percents = pd.concat([passing_network["Passes From v to >"],
    (passing_network.iloc[:,1:-1].div(passing_network["total_passes"].values,
         axis = 0)).multiply(100).round(2)], axis = 1)

passing_network_as_percents
```

```python
# Figure out how to make an edgelist with the top x connections 
cutoff_players = 2
edge_weighting_factor = 10
nplayers = len(passing_network_as_percents)
player_data_df = pd.DataFrame()
edgelist = pd.DataFrame()

for i in range(nplayers):
    player_to_last_name = []
    player_data_to = passing_network_as_percents.iloc[i,1:] #data on player being passed to
    player_from = passing_network_as_percents.iloc[i,0].split()[0] #last name name of player doing the passing

    #sort the values 
    player_data_to.sort_values(ascending=False, inplace=True)
    
    #extract last names 
    for player_name in player_data_to.iloc[:cutoff_players].index:
        player_to_last_name.append(player_name.split()[0])

    player_data_df["From"] = [player_from] * cutoff_players
    player_data_df["To"] = player_to_last_name
    player_data_df["value"] = player_data_to[:cutoff_players].values / edge_weighting_factor
    
    #append to higher dataframe
    edgelist = edgelist.append(player_data_df, ignore_index=True)
# player_data_df.head(cutoff_players)
edgelist
```

```python
#now to use networkx
import networkx as nx
G = nx.from_pandas_edgelist(edgelist,
                            source="From",
                            target="To",
                            edge_attr="value",
                            create_using=nx.DiGraph())
```

```python
# import pyvis
from pyvis.network import Network
# create vis network
net = Network(notebook=True, width=1000, height=600, directed=True)
# load the networkx graph
net.from_nx(G)

#specify node positions for visuals
formation = {"PICKFORD":(0,0),
             "WALKER":(-500,110),
             "SHAW":(500,110),
             "RICE":(0,200),
             "STONES":(-300,100),
             "MAGUIRE":(300,100),
             "HENDERSON":(-300,250),
             "KANE":(0,600),
             "SAKA":(-500,500),
             "FODEN":(500,500),
             "BELLINGHAM":(300,250)}


for i,node in enumerate(net.nodes):
    
    node.update({'size':50,
                 'physics':False,
                 'x':formation[node["label"]][0],
                 'y':formation[node["label"]][1]})
# show
net.force_atlas_2based()
net.show_buttons(filter_=['physics'])
net.show(f"../outputs/passing_network_vs_{opposition}.html")
```
